```python
import joypy 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\ritad\\Downloads\\Istanbul Weather Data.csv')
df.head()
```

```python
df['DateTime'] = pd.to_datetime(df['DateTime'])
df.head()
```

```python
df_novo = df.loc[(df['DateTime'] < '2019-01-01')]
df_novo.head()
```

```python
df_novo2 = df_novo[(df_novo['DateTime'] > '2017-12-31')]
df_novo2.tail()
```

```python
from matplotlib.lines import Line2D


cmap = plt.cm.coolwarm


df_novo2['Num_Month'] = df_novo2['DateTime'].dt.month
df_novo2['Month'] = df_novo2['DateTime'].dt.strftime('%B')
df_novo2.drop(columns = ['month', 'month_num'])

custom_lines = [Line2D([0], [0], color=cmap(1.), lw=4),
                Line2D([0], [0], color=cmap(.5), lw=4)]

fig, ax = joypy.joyplot(data = df_novo2,
                       by = 'Num_Month',
                       column=['MaxTemp', 'MinTemp'],
                       labels = ['Janeiro', 'Fevereiro', 'Maro', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro',
                       'Outubro', 'Novembro', 'Dezembro'],
                       color = ['red', 'white'],
                       legend = False,
                        background = 'grey',
                       alpha = .5)

fig.legend(custom_lines, ['Temperatura mxima', 'Temperatura mnima'], bbox_to_anchor=(0.5, 0), loc = 'upper center', ncol = 2, labelcolor = ['black', 'black'], 
           facecolor = 'grey', edgecolor = 'grey', fontsize = '12')

fig.suptitle('Temperaturas anuais em Istambul - 2018', fontname='Sans', size = 20, color = 'black',
            horizontalalignment='center', fontweight = 'bold')
fig.set_size_inches(10,6)
fig.patch.set_facecolor('grey')
plt.xlabel('Temperatura', size='12', fontname = 'Sans', fontweight = 'bold')
plt.xticks(fontsize = 12, fontname = 'Sans', fontweight = 'bold')
plt.tick_params(bottom=False)
plt.text(32, 1, '@rafaelf_lima', size = '8')
for x in ax:
    label = x.get_yticklabels()
    label[0].set_color('black')
    label[0].set_fontsize('12')
    label[0].set_fontweight('bold')

```
